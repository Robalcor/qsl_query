## Encuentra el n√∫mero de rese√±as de usuarios y la calificaci√≥n promedio para cada libro.

```python
query_rating = """
    SELECT 
        books.title AS title,
        AVG(ratings.rating) AS avg_rating,
        COUNT(ratings.rating) AS cnt_rating
    FROM
        books
        LEFT JOIN ratings ON books.book_id = ratings.book_id
    GROUP BY 
        books.title
    ORDER BY
        cnt_rating DESC;
"""
pd.io.sql.read_sql(query_rating, con = engine)
```

La calificaci√≥n promedio con pocas rese√±as no es tan confiable.

Entre m√°s calificaciones mejor y mas confiable es la calificaci√≥n promedio.

## Identifica la editorial que ha publicado el mayor n√∫mero de libros con m√°s de 50 p√°ginas (esto te ayudar√° a excluir folletos y publicaciones similares de tu an√°lisis).

```python
query_p50 = """
    SELECT 
        publishers.publisher AS publisher,
        COUNT(books.title) AS cnt_title
    FROM
        publishers
        LEFT JOIN books ON books.publisher_id = publishers.publisher_id
    WHERE 
        books.num_pages > 50
    GROUP BY 
        publishers.publisher
    ORDER BY
        cnt_title DESC
    LIMIT 1;    
"""
pd.io.sql.read_sql(query_p50, con = engine)
```

## Identifica al autor que tiene la m√°s alta calificaci√≥n promedio del libro: mira solo los libros con al menos 50 calificaciones.

```python
query_maxra = """
    SELECT 
        authors.author AS author,
        AVG(subquery.avg_rating) AS avg_rating
    FROM 
       authors
    JOIN (
        SELECT 
            books.author_id AS author_id,
            AVG(ratings.rating) AS avg_rating,
            COUNT(ratings.rating) AS cnt_rating
        FROM
            books
        JOIN ratings ON books.book_id = ratings.book_id
        GROUP BY 
            books.author_id
        ) AS subquery ON subquery.author_id = authors.author_id
    WHERE
        subquery.cnt_rating > 50
    GROUP BY
        authors.author
    ORDER BY
        avg_rating DESC
    LIMIT 1
"""
pd.io.sql.read_sql(query_maxra, con = engine)
```

## Encuentra el n√∫mero promedio de rese√±as de texto entre los usuarios que calificaron m√°s de 50 libros.

```python
query_us = """
    SELECT
        reviews.username,
        COUNT(reviews.text) AS cnt_reviews,
        AVG(COUNT(reviews.text)) OVER () AS promedio_general
    FROM 
        reviews
        JOIN (
            SELECT
                username,
                COUNT(rating) AS cnt_rating
            FROM 
                ratings
            GROUP BY
                username  
                ) AS subquery ON subquery.username = reviews.username
    WHERE
        subquery.cnt_rating > 50
    GROUP BY
        reviews.username    
"""
pd.io.sql.read_sql(query_us, con = engine)
```

Los usuarios que han calificado mas de 50 libro hacen en promedio 24.33 rese√±as. Podr√≠a decirse que rese√±an la mitad de los libros que califican.
